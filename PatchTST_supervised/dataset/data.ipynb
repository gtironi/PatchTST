```python
import pandas as pd

df = pd.read_csv("dog_raw.csv")
```

```python
df.columns
```

```python
df.drop(columns=['Subject', 'Position'], inplace=True)
df.columns
```

```python
"""
Estrutura esperada pelo código:

date, 0, 1, 2, 3, 4, ..., OT

sempre deve começar com date e terminar com OT
"""

# primeira coluna como date
df.columns.values[0] = "date"

# features como index
for index, col in enumerate(df.columns.values[1:]):
    print("-"*10)
    print(col, "->", index + 1)
    df.columns.values[index + 1] = index

#dummy apeans para rodar
df["OT"] = df.iloc[:, -1]  # ← ADICIONA COLUNA 'OT' COM OS MESMOS VALORES DA ÚLTIMA COLUNA

df.columns
```

```python
df = df[0:20000]

print(len(df))

df.to_csv("dog.csv", index=False)
```
