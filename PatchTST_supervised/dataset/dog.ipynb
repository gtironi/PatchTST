```python
import pandas as pd

df = pd.read_csv("dog_raw.csv")
```

```python
df.columns
```

```python
df.drop(columns=['Subject', 'Position'], inplace=True)
df.columns
```

```python
"""
Expected structure by the code:

date, 0, 1, 2, 3, 4, ..., OT

should always start with date and end with OT
"""

# first column as date
df.columns.values[0] = "date"

# features as index
for index, col in enumerate(df.columns.values[1:]):
    print(col, "->", index)
    df.columns.values[index + 1] = index

# dummy column just to run
df["OT"] = df.iloc[:, -1]  # same values as the last column (not used, dummy)

print("final structure")
df.columns
```
